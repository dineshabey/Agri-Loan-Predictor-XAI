# 2.Data Preprocessing


 Define the Mapping Dictionary for Months
* Why: Mapping Sinhala month names to English enables compatibility with ML tools like SHAP and Matplotlib.
* Result: Your dataset will have standardized column names recognized globally.




```python
# 2. Define English Translation Mapping
column_map = {
    'ණය වර්ගය': 'Loan_Type',
    'කල් පසු වී ඇත / නැත': 'Overdue_Status',
    'ගත් ක්‍රියා මාර්ගය': 'Action_Taken',
    'කෘ ප. නි. ස. නිලධාරීවරයෙකු ඇත/ නැත': 'Officer_Assigned',
    'කොට්ඨාසය': 'Division',
    'ණය මුදල': 'Loan_Amount',
    'ජනවාරි අයවීම්': 'Jan_Recovery',
    'පෙබරවාරි අයවීම්': 'Feb_Recovery',
    'මාර්තු අයවීම්': 'Mar_Recovery',
    'අප්‍රේල් අයවීම්': 'Apr_Recovery',
    'මැයි අයවීම්': 'May_Recovery',
    'ජුනි අයවීම්': 'Jun_Recovery',
    'ජූලි අයවීම්': 'Jul_Recovery',
    'අගෝස්තු අයවීම්': 'Aug_Recovery',
    'සැප්තැම්බර් අයවීම්': 'Sep_Recovery',
    'ඔක්තොබර් අයවීම්': 'Oct_Recovery',
    'නොවැම්බර් අයවීම්': 'Nov_Recovery',
    'දෙසැම්බර් අයවීම්': 'Dec_Recovery',
    'හිග ශේෂය': 'Outstanding_Balance',

}

# 3. Rename and Drop PII (Ethics Requirement)
df_translated = df.rename(columns=column_map)
df_translated = df_translated.drop(columns=['නම', 'ණය අංකය'], errors='ignore')

# 4. Map Sinhala values to English
val_map = {'නැත': 'No', 'ඇත': 'Yes' , 'විනිශ්‍ය සභා': 'Adjudication_Board'}
cols_to_fix = ['Overdue_Status', 'Action_Taken', 'Officer_Assigned']
for col in cols_to_fix:
    df_translated[col] = df_translated[col].map(val_map).fillna(df_translated[col])


```

```python
# 'මහ' refers to the 'Maha' season, and 'යල' refers to 'Yala' season in Sri Lankan agriculture.
loan_type_map = {
    '2025 මහ කන්නය': '2024_Maha_season'
}
df_translated['Loan_Type'] = df_translated['Loan_Type'].map(loan_type_map).fillna(df_translated['Loan_Type'])
```

```python
# 1. EXTRACT UNIQUE NAMES
# Why: To identify all unique geographical areas without duplicates for translation.
# Result: A clean list of every unique division in your Sri Lankan dataset.
unique_divisions = df_translated['Division'].unique().tolist()

# Remove NaN values before sorting
unique_divisions = [div for div in unique_divisions if pd.notna(div)]

# 2. SORT AND PRINT
# Why: Sorting makes it easier to verify the names in your report.
unique_divisions.sort()

print(f"Total Unique Divisions: {len(unique_divisions)}")
print("--- Full Division List ---")
for div in unique_divisions:
    print(div)
```

```python
from google.colab import userdata
from groq import Groq
groq_api_key = userdata.get("GROQ_API_KEY")
client = Groq(api_key=groq_api_key)
import time
```

```python
def translate_division_llm(sinhala_name):
    prompt = f"""
Translate the following Sri Lankan administrative division name
from Sinhala to proper English.

Return ONLY the English name.
Do not explain.

Sinhala: {sinhala_name}
"""
    response = client.chat.completions.create(
          model="llama-3.1-8b-instant",  # updated working model
          messages=[{"role": "user", "content": prompt}],
          temperature=0.0
    )
    return response.choices[0].message.content.strip()
```

```python
division_map = {}   # Sinhala → English
logs = []

for div in unique_divisions:
    if div not in division_map:
        try:
            english = translate_division_llm(div)
            division_map[div] = english

            logs.append({
                "Sinhala": div,
                "English": english
            })

            print(f"{div}  →  {english}")
            time.sleep(0.5)  # be polite to free API

        except Exception as e:
            division_map[div] = "Unknown"
            print(f"ERROR for {div}: {e}")
```

```python
df_translated["Division"] = (
    df_translated["Division"]
    .map(division_map)
    .fillna("Unknown")
)
```
